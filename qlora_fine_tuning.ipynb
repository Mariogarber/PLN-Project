# PROPUESTA
El proyecto se basa en la implementación de una estrategia de fine-tuning QLoRa en una arquitectura Seq2Seq Transformer. Para optimizar el rendimiento en tareas de detección o mitigación de toxicidad, se procederá al ajuste ponderado de la función de pérdida, asignando una mayor relevancia a los tokens correspondientes a las palabras clave (keywords) y segmentos (spans) identificados como tóxicos en los datasets.
