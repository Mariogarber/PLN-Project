# ☣️ PAN 2025 de Multilingual Text Detoxification ☣️

### 1. Idea Inicial (Metodología Descartada)

En un principio se evaluó una arquitectura de dos etapas con dos modelos para cada función:

- **Modelo 1 (Enmascaramiento):** Entrenado con *datasets* de toxicidad para identificar y **enmascarar** los segmentos problemáticos.  
- **Modelo 2 (Desintoxicación):** Encargado de tomar el texto enmascarado y generar la versión final no tóxica.

**Razón del Cambio:** Esta estrategia limitaba la capacidad del sistema para reformular oraciones completas, un aspecto fundamental en el reto. La rigidez del proceso de enmascaramiento y la dificultad para interpretar sus resultados impedían una reescritura fluida y contextual, lo que derivaba en salidas poco naturales o incompletas.

### 2. Enfoque Actual: Ponderación de Pérdida Única

La metodología actual se enfoca en una única estrategia, integrando la detección y mitigación en un solo modelo:

| Aspecto | Especificación | Detalle |
| :--- | :--- | :--- |
| Modelo Base | MT5-Base | Es una elección adecuada porque la arquitectura Transformer Seq2Seq es un estándar consolidado para tareas de reescritura generativa. Además, su preentrenamiento multilingüe resulta especialmente valioso para cubrir los 15 idiomas planteados en el reto.|
| **Técnica de Entrenamiento** | **QLoRa** (Quantized Low-Rank Adaptation) | Permite el *fine-tuning* eficiente de modelos grandes con recursos limitados. |
| **Innovación Central** | **Ajuste Ponderado de la Función de Pérdida** | Los *tokens* correspondientes a la toxicidad (keywords y spans) reciben un **peso de contribución (loss weight)** más alto. Con el objetivo de penalizar más severamente los errores cometidos en esas partes específicas del texto para mejorar la identificación y mitigación.|
